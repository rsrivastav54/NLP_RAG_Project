```python
def split_text(text):
    # Split by newlines first
    lines = text.split('\n')
    # Now split each line by '.', then strip each part to remove leading/trailing spaces
    split_texts = [sentence.strip() for line in lines for sentence in line.split('.') if sentence]
    return split_texts


# Convert the Context column to string type
sample_df.loc[:, 'Context'] = sample_df['Context'].astype(str)

# Apply the function to the column
sample_df['Context'] = sample_df['Context'].apply(split_text)

print(sample_df['Context'])
```
